```python
# Fix output folder
output_folder_test = "example_simulation_plots/"
```

# Starting the simulation:
Now We simulate the Galactic Sky.

Some details:
- In this case all the outputs are saved to the variable arr along with npz and png files in the output folder given above
- For the test distribution model we use "exponential" along with 10000 sources being simulated.
- Note that the r(min,max) and z(min, max) values can be changed based on requirement.
- For more details on how the outputs change, see the article: https://iopscience.iop.org/article/10.3847/1538-4357/ad2a5e

```python
#For now this code will save all the source position plots (2 aitoffs and 1 3d) to the directory given below.
# If plots are not required just remove the plt_dir,plot_aitoff_dir_icrs and plot_aitoff_dir_gal lines.

start_time = time.time()

arr=simulate_nu_g.simulate_positions(number_sources_used  = int(1e4),
                                     distribution_model    ="exponential", #"modified_exponential",
                                     # FOLLOWING VALUES CAN BE CHANGED AS DESIRED
                                     r_min                = 1e-3,
                                     r_max                = 1e3, 
                                     z_min                = 1e-6,
                                     z_max                = 300,
                                     #NOTE GIVING THE FILE LOACTIONS BELOW WILL MAKE PLOTS!
                                     make_pdf_plot_location= output_folder_test,
                                     plot_dir             = output_folder_test, 
                                     plot_aitoff_dir_icrs = output_folder_test,
                                     plot_aitoff_dir_gal  = output_folder_test,
                                     output_file          = output_folder_test)


print("--- %s seconds ---" %(time.time() - start_time))
```

```python
#To see the contents of the saved variable uncomment the following:
#print(arr)
```

```python
# In the event you want to retrive the saved coordinates from the npz files you can do the following:

simulated_sources = np.load(output_folder_test+"simulated_10000_source_coordinates.npz")
print("Numpy array files derived from saved File: ",simulated_sources.files)
print("\n - r value Read from saved File: \n",simulated_sources['r'])
print("\n - r value Read using arr Variable from this code: \n",arr[0][0])
# r,phi z are GALACTO-CENTRIC Coords while
# gal_l gal_b and gal_d are GALACTIC Coords (distances are in kpc)
```

# Next, we simulate the Fluxes:
Some details regarding this:
- The simulated coordinates from above are used for this portion. If you have known coordinates simulated or observed from any other source, they can be used here too as long as the format is in array[arr[r],arr[phi],arr[z]]
- In this case a LogNormal distribution is used for both the neutrino and Gamma-ray cases, along with the other relevant parameters as shown below.
- Note that both neutrino and gamma-ray fluxes are simulated independently in this test case scenario.
- For more details about different type of Flux simulations, see: https://iopscience.iop.org/article/10.3847/1538-4357/ad2a5e

```python
# USE SIMULATED DISTANCES TO GIVE FLUX VALUES

arr_with_flux = simulate_nu_g.Get_flux_from_positions(galcentric_coords_r_phi_z   = arr[0],
                                                      nu_method_used              = "LogNormal", #Neutrino
                                                      gamma_ray_method_used       = "LogNormal", #Gamma_ray
                                                      diffuse_nu_flux_given       = 2.18e-15, # Tev-1cm-2s-1 Neutrino Isotropic flux
                                                      diffuse_gamma_flux_given    = 7e-17, # Tev-1cm-2s-1 Gamma_ray Isotropic flux
                                                      print_output                = False,
                                                      full_path                   = None, #"./ default.npy",
                                                      index_nu_given              = 2.7, #Neutrino
                                                      index_gamma_given           = 2.7, #Gamma_ray
                                                      nu_ref_energy               = 100.0, #NEUTRINO reference energy
                                                      simulate_gamma_ray_frm_nu   = False,
                                                      simulate_nu_frm_gamma       = False,
                                                      mean_luminosity_nu          = None,
                                                      mean_luminosity_gamma       = 1e35,
                                                      stdev_sigma_L_nu            = 0.6,
                                                      stdev_sigma_L_gamma         = 0.6)

```

```python
print("Diffuse_flux is 2.14e-15")
print("Simulated Total is ",arr_with_flux[1].sum())
arr_with_flux[-1:]
```

# Bonus Plot:
- Lets plot the simulated neutrino sources in a color map in an aitoff projection.
- A similar code was used in Chapter 3 of the following work:
  https://open.clemson.edu/all_dissertations/3412/

```python
from astropy.coordinates import SkyCoord
import matplotlib.colors as colors
from astropy import units as u

gal_coords1 = arr[2]
ra = gal_coords1.l.wrap_at(180 * u.deg).radian
dec = gal_coords1.b.wrap_at(180 * u.deg).radian#z_list = arr_with_flux[1]/sum(arr_with_flux[1])
z_list = 10**arr_with_flux[1]
z_list=z_list/sum(z_list)

coords = gal_coords1


fig = plt.figure(figsize=(20,10),dpi=250)
ax= fig.add_subplot(111, projection='aitoff')
ax.set_facecolor("Ivory")
image = plt.scatter(coords.l.wrap_at('180d').radian, coords.b.radian, c=z_list, cmap='plasma', norm=colors.LogNorm(vmax=np.nanmax(z_list)))
#plt.legend(loc="lower right")

plt.xlabel('l')
plt.ylabel('b.')
plt.grid(True)
plt.colorbar(image, spacing='uniform',label="Neutrino Flux",ticks=[0,1])

```
